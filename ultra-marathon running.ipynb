```python
# is it slower to run in summer than in winter? 

# spring 3-5 
# summer 6 - 8
# fall 9 -11 
# winter 12 - 2
# splitting between two decimals 
```

```python
marathon ['race_month'] = marathon['race_day'].str.get(1).astype(int)
```

```python
marathon.head(5)
```

```python
marathon['race_season'] = marathon['race_month'].apply (lambda x: 'Winter' if x > 11 else 'Fall' if x > 8 else 
                                                        'Summer' if x > 5 else  'Spring' if x > 2 else 'Winter')
```

```python
marathon.head(5)
```
